Read images to covert it to 3d array or array of 2d images in grayscale.

[Help taken from](https://stackoverflow.com/a/49070833)

```python
# Source code is taken and edited from [Stack Overflow](https://stackoverflow.com/a/49070833)

from PIL import Image
import sys
import os
import csv

#Useful function
def createFileList(myDir, format='.jpg'):
  fileList = []
  print(myDir)
  for root, dirs, files in os.walk(myDir, topdown=False):
      for name in files:
          if name.endswith(format):
              fullName = os.path.join(root, name)
              fileList.append(fullName)
  return fileList

# Setup image dataset for test
myFileList = createFileList('/content/ML1/')

for file in myFileList:
    print(file)
    img_file = Image.open(file)
    # img_file.show()

    # get original image parameters...
    width, height = img_file.size
    format = img_file.format
    mode = img_file.mode

    # Make image Greyscale
    img_grey = img_file.convert('L')
    #img_grey.save('result.png')
    #img_grey.show()

    # Save Greyscale values
    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    value = value/255.0
    #value = value.flatten()
    #print(value)
    #with open("img_pixels.csv", 'a') as f:
        #writer = csv.writer(f)
        #writer.writerow(value)
    x_train.append(value)

# Setup data for train
myFileList = createFileList('/content/ML2/')

for f in myFileList:
    print("File : ",f)
    img_file = Image.open(f)
    # img_file.show()

    # get original image parameters...
    width, height = img_file.size
    format = img_file.format
    mode = img_file.mode

    # Make image Greyscale
    img_grey = img_file.convert('L')
    #img_grey.save('result.png')
    #img_grey.show()

    # Save Greyscale values
    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    value = value/255.0
    #value = value.flatten()
    # print(value)
    #with open("img_pixels.csv", 'a') as f:
        #writer = csv.writer(f)
        #writer.writerow(value)

    x_train.append(value)

x_train
```

## Define and compile the first model

```python
from keras.backend import batch_normalization
import keras
from keras.models import Sequential
from keras.layers import Conv2D, Lambda, MaxPooling2D
from keras.layers import Dense, Dropout, Flatten
from  keras.layers.normalization.batch_normalization import BatchNormalization

from keras.preprocessing.image import ImageDataGenerator

from keras.utils.np_utils import to_categorical


model=Sequential()
   
model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu", input_shape=(28,28,1)))
model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())

model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())    

model.add(Conv2D(filters=256, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
    
model.add(Flatten())
model.add(Dense(512,activation="relu"))
    
model.add(Dense(6,activation="softmax"))
    
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

## Train the first model

```python
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
model.fit(x_train, y_train, batch_size=32, epochs=10, validation_split=0.2)
```

## Setup for the second model to transfer knowledge
Uses MNISt dataset to learn handwrtten characters.

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

Remove the last layer from model and a new layer of size 10 to classify digits.

```python
model.pop()
model.add(Dense(10,activation="softmax"))
```

Freeze the layers except last and compile.

```python
for layer in model.layers[:-1]:
  layer.trainable = False


model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), optimizer="adam", metrics=["accuracy"])
```

Get the summary of model.

```python
model.summary()
```

Train the new model.

```python
model.fit(x_train, y_train, epochs=5)
```

Evalute the model.

```python
model.evaluate(x_test,  y_test, verbose=2)
```

Our new model gives accuracy of 82.84% on MNIST dataset.
